## Models

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="qwen/qwen3-32b",
    model_provider="groq",
    temperature=0,
)

r = model.invoke("Hello, how are you?")
r
```

init_chat_model() provides an unified interface across providers, you can switch by just changing the model and model_providers.

There are other individual provider-specific classes like ChatGroq(), ChatOpenAI(), ChatGoogleGenerativeAI() etc, if using this it is hard to switch between providers. So, for large scale projects it is suggested to use init_chat_models().

## Streaming

Most models can stream their output content while it is being generated. Streaming improves the User Experience.

Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
message = 'What is Agentic AI?'

for chunk in model.stream(message):
    print(chunk.text, end='', flush=True)
```

## Batching

Batching a collection of independent requests to a model can significantly improve performance & reduce costs as the processing can be done in parallel.

```python
responses = model.batch([
    "What is AI?",
    "Explain the concept of reinforcement learning.",
    "I need to buy new laptop, can you suggest some good options?",
],config={
        'max_concurrency': 5,  # Limit to 5 parallel calls
    }
)

for response in responses:
    print(response)
```
